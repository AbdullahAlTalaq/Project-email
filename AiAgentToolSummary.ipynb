```python
from smolagents import CodeAgent, DuckDuckGoSearchTool, HfApiModel, tool
import requests
from bs4 import BeautifulSoup
@tool
def my_tool(arg1:str) ->str:
  """
  A tool that make a summary of website page.
  Args:
    arg1: The url of the website page.
  Returns:
    The summary of the website page.

  """
  try:
    response = requests.get(arg1)
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup.get_text()
  except Exception as e:
    return f"Error {e}"


model = HfApiModel(max_tokens=2096,
temperature=0.5,
model_id='Qwen/Qwen2.5-Coder-32B-Instruct',# it is possible that this model may be overloaded
custom_role_conversions=None,)
agent = CodeAgent(model=model,tools=[my_tool])
agent.run("make a summary of this webpage 'https://www.geeksforgeeks.org/python-for-data-science/' ")
```
